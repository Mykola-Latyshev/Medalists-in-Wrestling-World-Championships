```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df_women_final = pd.read_csv('df_women_final.csv')
df_men_final = pd.read_csv('df_men_final.csv')
```

```python
# EDA 
# rank/target/target group/target medalist
```

```python
# target rank
```

```python
# count target for each rank
df_fig_target2 = df_men_final.groupby(['target_rank'])['Name'].count().reset_index(name='counts')
plt.figure(figsize = (10,7))
plt.bar(df_fig_target2['target_rank'], df_fig_target2['counts'])

font1 = {'family':'serif','color':'black','size':14}

plt.xlabel('Rank',  fontdict = font1)
plt.ylabel('The number of athletes',  fontdict = font1)
plt.savefig('rank.png')
```

```python
# target group
```

```python
# count target group
df_fig_target1 = df_men_final.groupby(['target_group'])['Name'].count().reset_index(name='counts')

plt.figure(figsize = (8,5))
plt.bar(df_fig_target1['target_group'], df_fig_target1['counts'])

font1 = {'family':'serif','color':'black','size':12}

plt.xlabel('Target group : 1 - medalist; 2 - medium; 3 - outsider', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)
```

```python
# target medalist
```

```python
# count target medalist
df_fig_target1 = df_men_final.groupby(['target_medalist'])['Name'].count().reset_index(name='counts')
plt.figure(figsize = (7,5))
plt.bar(df_fig_target1['target_medalist'], df_fig_target1['counts'], width=0.45, color ='teal')

font1 = {'family':'serif','color':'black','size':14}
plt.xlabel('Target : 1 - medalist; 0 - no medalist (medium and outsider)', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)
```

```python
# Count Percentage of medalists
print('Percentage of medalists, %:', (len(df_men_final.loc[df_men_final.target_medalist == 1])/len(df_men_final))*100)
```

```python
# EDA of age (hist)
```

```python
#sns.color_palette('pastel')

f, (ax_box1, ax_box2, ax_hist) = plt.subplots(nrows = 3, sharex=True, 
                                              gridspec_kw={"height_ratios": [0.13,0.13,1.2], 'width_ratios' : [1]},
                                             figsize=(9,8))
sns.boxplot(df_men_final.loc[df_men_final.target_medalist == 0]['Age_final'], color ='mediumblue', ax = ax_box1)
sns.boxplot(df_men_final.loc[df_men_final.target_medalist == 1]['Age_final'], color ='yellowgreen', ax = ax_box2)

sns.histplot(df_men_final.loc[df_men_final.target_medalist == 0], x='Age_final', stat="density", 
             element="step", bins = 20, color ='mediumblue', linewidth=2.5, ax = ax_hist)

sns.histplot(df_men_final.loc[df_men_final.target_medalist == 1], x='Age_final', stat="density", 
             element="step", color ='yellowgreen', bins =17, linewidth=2.5, ax = ax_hist)

f.subplots_adjust(hspace=0.05)

ax_box1.set(xlabel='')
ax_box2.set(xlabel='')

font1 = {'family':'serif','color':'black','size':16}

ax_hist.set_xlabel('Age', fontdict = font1)
ax_hist.set_ylabel('Density', fontdict = font1)
ax_hist.tick_params(axis='x', labelsize=12)
ax_hist.tick_params(axis='y', labelsize=12)
ax_hist.locator_params(axis='x', nbins=24)

f.savefig('hist_box_age.png')
```

```python
# the second example
# age of athletes 
plt.figure(figsize = (10,7))
plt.hist(df_men_final.loc[df_men_final.target_medalist == 0]['Age_final'], 
         bins=20, density=True, histtype= 'step', linewidth=2.5)
plt.hist(df_men_final.loc[df_men_final.target_medalist == 1]['Age_final'], 
         bins=17, density=True, histtype= 'barstacked')
```

```python
#Month_BirthDay
df_fig_mounth_birth = df_men_final.groupby(['Month_BirthDay'])['Name'].count().reset_index(name='counts')
plt.figure(figsize = (7,5))
plt.bar(df_fig_mounth_birth['Month_BirthDay'], df_fig_mounth_birth['counts'])

font1 = {'family':'serif','color':'black','size':14}
plt.xlabel('Month', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)

plt.savefig('03_dist_month_bd_m.png')
```

```python
# correlation between age data and target
age_col = ['Age_final', 'Month_BirthDay', 'Quarter_REA', 'Half_REA', 'target_rank', 'target_group']
corr_matrix = df_men_final.loc[:,age_col].corr()

plt.figure(figsize = (7,5))
sns.heatmap(corr_matrix, annot=True, fmt='.2g', cmap= 'coolwarm')
```

```python
# correlation between cadets/junior results and target
cadet_junior_col = ['Cadet_Win', 'Cadet_Success', 'Cadet_Part',
                    'Junior_Win', 'Junior_Success', 'Junior_Part',
                    'target_rank', 'target_group']
corr_matrix = df_men_final.loc[:,cadet_junior_col].corr()

plt.figure(figsize = (7,5))
sns.heatmap(corr_matrix, annot=True, fmt='.2g', cmap= 'coolwarm')
```

```python
# correlation between cadets/junior results and target
senior_col = ['Senior_Win_CC', 'Senior_Success_CC', 'Senior_Part_CC',
                'Senior_Win_WC', 'Senior_Success_WC', 'Senior_Part_WC',
                'target_rank', 'target_group', 'target_medalist']
corr_matrix = df_men_final.loc[:,senior_col].corr()

plt.figure(figsize = (7,5))
sns.heatmap(corr_matrix, annot=True, fmt='.2g', cmap= 'coolwarm')

plt.savefig('04_matrix_corr_senior_m.png')
```

```python
#Cadet_success
#df_men_final.loc[df_men_final.target_medalist == 0]
df_fig_cadets_part_med = df_men_final.loc[df_men_final.target_medalist == 1]. \
                        groupby(['Cadet_Part'])['Name'].count().reset_index(name='counts')

df_fig_cadets_part_no_med = df_men_final.loc[df_men_final.target_medalist == 0]. \
                        groupby(['Cadet_Part'])['Name'].count().reset_index(name='counts')

plt.figure(figsize = (8,6))
offset = 0.2
plt.bar(df_fig_cadets_part_med['Cadet_Part'] - offset, 
        (df_fig_cadets_part_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 1])*100),
         width=0.34, color ='teal')

plt.bar(df_fig_cadets_part_no_med['Cadet_Part'] + offset, 
        (df_fig_cadets_part_no_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 0])*100), 
        width=0.34, color ='olive')

font1 = {'family':'serif','color':'black','size':14}
plt.xlabel('The number participation of athletes on cadets competitions', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)
plt.legend(['Medalist', 'No medalist'], fontsize = 'large')

plt.savefig('05_dist_count_cadet_part_m.png')
```

```python
#Junior_Success
#df_men_final.loc[df_men_final.target_medalist == 0]
df_fig_junior_suc_med = df_men_final.loc[df_men_final.target_medalist == 1]. \
                        groupby(['Junior_Success'])['Name'].count().reset_index(name='counts')

df_fig_junior_suc_no_med = df_men_final.loc[df_men_final.target_medalist == 0]. \
                        groupby(['Junior_Success'])['Name'].count().reset_index(name='counts')

plt.figure(figsize = (8,6))
offset = 0.2
plt.bar(df_fig_junior_suc_med['Junior_Success'] - offset, 
        (df_fig_junior_suc_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 1])*100),
         width=0.3)

plt.bar(df_fig_junior_suc_no_med['Junior_Success'] + offset, 
        (df_fig_junior_suc_no_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 0])*100), 
        color = 'green', width=0.3)


font1 = {'family':'serif','color':'black','size':14}
plt.xlabel('The number success of athletes on juniors competitions', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)
plt.legend(['Medalist', 'No medalist'], fontsize = 'large')

plt.savefig('06_dist_count_cadet_part_m.png')
```

```python
#Senior_Part_CC
#df_fig_senior_suc_cc = df_m_new2.groupby(['Senior_Success_CC'])['Name'].count().reset_index(name='counts')
#plt.figure(figsize = (7,5))
#plt.bar(df_fig_senior_part_cc['Senior_Success_CC'], df_fig_senior_part_cc['counts'])

#plt.savefig('dist_count_senior_coninent_part_m.png')

df_fig_senior_suc_med = df_men_final.loc[df_men_final.target_medalist == 1]. \
                        groupby(['Senior_Success_WC'])['Name'].count().reset_index(name='counts')

df_fig_senior_suc_no_med = df_men_final.loc[df_men_final.target_medalist == 0]. \
                        groupby(['Senior_Success_WC'])['Name'].count().reset_index(name='counts')

plt.figure(figsize = (7,5))
offset = 0.2
plt.bar(df_fig_senior_suc_med['Senior_Success_WC'] - offset, 
        (df_fig_senior_suc_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 1])*100),
         width=0.3)

plt.bar(df_fig_senior_suc_no_med['Senior_Success_WC'] + offset, 
        (df_fig_senior_suc_no_med['counts']/len(df_men_final.loc[df_men_final.target_medalist == 0])*100), 
        color = 'green', width=0.3)

font1 = {'family':'serif','color':'black','size':14}
plt.xlabel('The number success of athletes on world championships before the analyzed competition', fontdict = font1)
plt.ylabel('The number of athletes', fontdict = font1)
plt.legend(['Medalist', 'No medalist'], fontsize = 'large')

plt.savefig('07_world_count_succ_m.png')
```
